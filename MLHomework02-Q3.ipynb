```python
import os
import zipfile
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models, regularizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split
import pandas as pd
from tabulate import tabulate

# 設定參數
EPOCHS = 20
BATCH_SIZE = 32
IMG_HEIGHT = 150
IMG_WIDTH = 150
VALIDATION_SPLIT = 0.2

# 1. 載入和解壓ZIP檔
zip_path = r"C:\ML\archive.zip"
extract_dir = r"C:\ML\extracted_data"

# 如果資料夾不存在，就建立它
if not os.path.exists(extract_dir):
    os.makedirs(extract_dir)

# 解壓縮檔案
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

# 假設解壓後的目錄結構，依實際情況調整
train_dir = os.path.join(extract_dir, "train")
test_dir = os.path.join(extract_dir, "test")

# 2. 資料預處理與資料增強
# 針對訓練資料的資料增強
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
    validation_split=VALIDATION_SPLIT  # 分出一部分做為驗證集
)

# 針對測試資料，只進行重新縮放
test_datagen = ImageDataGenerator(rescale=1./255)

# 載入訓練資料
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training'  # 指定這是訓練集
)

# 載入驗證資料
validation_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation'  # 指定這是驗證集
)

# 載入測試資料
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

# 獲取類別數量
num_classes = len(train_generator.class_indices)
print(f"類別數量: {num_classes}")
print(f"類別對應: {train_generator.class_indices}")

# 3. 建立基礎CNN模型函數
def create_base_model():
    model = models.Sequential([
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_HEIGHT, IMG_WIDTH, 3)),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Flatten(),
        layers.Dense(512, activation='relu'),
        layers.Dense(num_classes, activation='softmax')
    ])
    return model

# 4. 實作不同的正規化技術
# 4.1 基礎模型（無正規化）
def create_baseline_model():
    model = create_base_model()
    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

# 4.2 使用Dropout的模型
def create_dropout_model():
    model = models.Sequential([
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_HEIGHT, IMG_WIDTH, 3)),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Flatten(),
        layers.Dropout(0.5),  # 添加Dropout
        layers.Dense(512, activation='relu'),
        layers.Dropout(0.5),  # 添加Dropout
        layers.Dense(num_classes, activation='softmax')
    ])
    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

# 4.3 使用Batch Normalization的模型
def create_batchnorm_model():
    model = models.Sequential([
        layers.Conv2D(32, (3, 3), padding='same', input_shape=(IMG_HEIGHT, IMG_WIDTH, 3)),
        layers.BatchNormalization(),  # 添加Batch Normalization
        layers.Activation('relu'),
        layers.MaxPooling2D((2, 2)),
        
        layers.Conv2D(64, (3, 3), padding='same'),
        layers.BatchNormalization(),  # 添加Batch Normalization
        layers.Activation('relu'),
        layers.MaxPooling2D((2, 2)),
        
        layers.Conv2D(128, (3, 3), padding='same'),
        layers.BatchNormalization(),  # 添加Batch Normalization
        layers.Activation('relu'),
        layers.MaxPooling2D((2, 2)),
        
        layers.Flatten(),
        layers.Dense(512),
        layers.BatchNormalization(),  # 添加Batch Normalization
        layers.Activation('relu'),
        layers.Dense(num_classes, activation='softmax')
    ])
    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

# 4.4 使用L1正規化的模型
def create_l1_model():
    model = models.Sequential([
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_HEIGHT, IMG_WIDTH, 3)),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Flatten(),
        layers.Dense(512, activation='relu', kernel_regularizer=regularizers.l1(0.001)),  # 添加L1正規化
        layers.Dense(num_classes, activation='softmax')
    ])
    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

# 4.5 使用L2正規化的模型
def create_l2_model():
    model = models.Sequential([
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_HEIGHT, IMG_WIDTH, 3)),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Flatten(),
        layers.Dense(512, activation='relu', kernel_regularizer=regularizers.l2(0.001)),  # 添加L2正規化
        layers.Dense(num_classes, activation='softmax')
    ])
    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

# 4.6 綜合技術模型 (Dropout + BatchNorm + L2 正規化)
def create_combined_model():
    model = models.Sequential([
        # 第一個卷積層區塊
        layers.Conv2D(32, (3, 3), padding='same', kernel_regularizer=regularizers.l2(0.0005), input_shape=(IMG_HEIGHT, IMG_WIDTH, 3)),
        layers.BatchNormalization(),
        layers.Activation('relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),
        
        # 第二個卷積層區塊
        layers.Conv2D(64, (3, 3), padding='same', kernel_regularizer=regularizers.l2(0.0005)),
        layers.BatchNormalization(),
        layers.Activation('relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),
        
        # 第三個卷積層區塊
        layers.Conv2D(128, (3, 3), padding='same', kernel_regularizer=regularizers.l2(0.0005)),
        layers.BatchNormalization(),
        layers.Activation('relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),
        
        # 全連接層
        layers.Flatten(),
        layers.Dense(512, kernel_regularizer=regularizers.l2(0.0005)),
        layers.BatchNormalization(),
        layers.Activation('relu'),
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation='softmax')
    ])
    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

# 5. 訓練並評估模型
def train_and_evaluate_model(model, model_name):
    print(f"\n訓練模型: {model_name}")
    
    # 訓練模型
    history = model.fit(
        train_generator,
        epochs=EPOCHS,
        validation_data=validation_generator,
        verbose=1
    )
    
    # 評估模型在測試集上的表現
    test_loss, test_acc = model.evaluate(test_generator)
    print(f"{model_name} - 測試集準確率: {test_acc:.4f}")
    
    # 評估模型在訓練集上的表現
    train_loss, train_acc = model.evaluate(train_generator)
    print(f"{model_name} - 訓練集準確率: {train_acc:.4f}")
    
    return history, train_acc, test_acc

# 建立和訓練所有模型
models_to_train = [
    ("基礎模型 (無正規化)", create_baseline_model()),
    ("Dropout 模型", create_dropout_model()),
    ("Batch Normalization 模型", create_batchnorm_model()),
    ("L1 正規化模型", create_l1_model()),
    ("L2 正規化模型", create_l2_model()),
    ("綜合技術模型 (Dropout+BatchNorm+L2)", create_combined_model())
]

results = []

for model_name, model in models_to_train:
    history, train_acc, test_acc = train_and_evaluate_model(model, model_name)
    results.append({
        "name": model_name,
        "history": history,
        "train_acc": train_acc,
        "test_acc": test_acc,
        "overfitting_gap": train_acc - test_acc
    })

# 繪製結果比較圖
def plot_results(results):
    # 準確率比較
    names = [result["name"] for result in results]
    train_accs = [result["train_acc"] for result in results]
    test_accs = [result["test_acc"] for result in results]
    
    x = np.arange(len(names))
    width = 0.35
    
    fig, ax = plt.subplots(figsize=(14, 8))
    rects1 = ax.bar(x - width/2, train_accs, width, label='訓練集準確率')
    rects2 = ax.bar(x + width/2, test_accs, width, label='測試集準確率')
    
    # 添加文字和標籤
    ax.set_ylabel('準確率')
    ax.set_title('不同正規化技術的準確率比較')
    ax.set_xticks(x)
    ax.set_xticklabels(names, rotation=45, ha='right')
    ax.legend()
    
    # 在每個長條上方顯示數值 (四位小數)
    def autolabel(rects):
        for rect in rects:
            height = rect.get_height()
            ax.annotate(f'{height:.4f}',
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 3),  # 3 points vertical offset
                        textcoords="offset points",
                        ha='center', va='bottom')
    
    autolabel(rects1)
    autolabel(rects2)
    
    fig.tight_layout()
    plt.savefig('regularization_comparison.png')
    plt.show()
    
    # 繪製訓練歷史
    fig, axs = plt.subplots(2, 3, figsize=(18, 10))
    axs = axs.ravel()
    
    for i, result in enumerate(results):
        history = result["history"]
        axs[i].plot(history.history['accuracy'])
        axs[i].plot(history.history['val_accuracy'])
        axs[i].set_title(result["name"])
        axs[i].set_ylabel('準確率')
        axs[i].set_xlabel('Epoch')
        axs[i].legend(['訓練', '驗證'], loc='lower right')
        
        # 設定y軸格式為四位小數
        axs[i].yaxis.set_major_formatter(plt.FormatStrFormatter('%.4f'))
    
    plt.tight_layout()
    plt.savefig('training_history.png')
    plt.show()

# 繪製結果
plot_results(results)

# 找出最佳模型
best_model_idx = np.argmax([result["test_acc"] for result in results])
best_model_name = results[best_model_idx]["name"]
best_model_test_acc = results[best_model_idx]["test_acc"]
best_model_train_acc = results[best_model_idx]["train_acc"]

# 顯示結果表格
def display_results_table(results):
    # 創建表格數據
    table_data = []
    for result in results:
        table_data.append([
            result["name"],
            f"{result['train_acc']:.4f}",
            f"{result['test_acc']:.4f}",
            f"{result['overfitting_gap']:.4f}"
        ])
    
    # 按測試集準確率降序排序
    table_data.sort(key=lambda x: float(x[2]), reverse=True)
    
    # 建立表格
    headers = ["模型", "訓練集準確率", "測試集準確率", "過擬合差距"]
    
    # 使用pandas建立表格並輸出
    df = pd.DataFrame(table_data, columns=headers)
    print("\n=== 模型效能比較表 ===")
    print(tabulate(df, headers='keys', tablefmt='pipe', showindex=False))
    
    # 找出最佳模型和過擬合最小的模型
    best_model_row = df.iloc[0]
    min_overfitting_row = df.iloc[df['過擬合差距'].astype(float).argmin()]
    
    print("\n=== 重要指標 ===")
    print(f"最佳模型 (測試集準確率): {best_model_row['模型']} - {best_model_row['測試集準確率']}")
    print(f"過擬合最小的模型: {min_overfitting_row['模型']} - 差距: {min_overfitting_row['過擬合差距']}")
    
    return df

# 嘗試導入pandas和tabulate
try:
    # 顯示表格結果
    results_df = display_results_table(results)
    
    # 保存結果到CSV檔案，確保四位小數
    pd.options.display.float_format = '{:.4f}'.format
    results_df.to_csv('regularization_results.csv', index=False, float_format='%.4f')
    print("\n結果已保存到 'regularization_results.csv'")
    
except ImportError:
    # 如果沒有安裝pandas或tabulate，使用簡單的表格輸出
    print("\n=== 模型效能比較表 ===")
    print("{:<35} {:<18} {:<18} {:<18}".format("模型", "訓練集準確率", "測試集準確率", "過擬合差距"))
    print("-" * 90)
    
    # 按測試集準確率降序排序
    sorted_results = sorted(results, key=lambda x: x["test_acc"], reverse=True)
    
    for result in sorted_results:
        print("{:<35} {:<18.4f} {:<18.4f} {:<18.4f}".format(
            result["name"], 
            result["train_acc"], 
            result["test_acc"], 
            result["overfitting_gap"]
        ))
    
    # 找出最佳模型
    best_model = max(results, key=lambda x: x["test_acc"])
    min_overfit_model = min(results, key=lambda x: x["overfitting_gap"])
    
    print("\n=== 重要指標 ===")
    print(f"最佳模型 (測試集準確率): {best_model['name']} - {best_model['test_acc']:.4f}")
    print(f"過擬合最小的模型: {min_overfit_model['name']} - 差距: {min_overfit_model['overfitting_gap']:.4f}")

# 產生完整的比較報告
def generate_detailed_report():
    print("\n=== 詳細比較報告 ===")
    
    # 表格資料
    report_data = []
    
    for result in results:
        # 從訓練歷史中獲取更多指標
        history = result["history"].history
        
        # 獲取最後一個epoch的驗證準確率
        final_val_acc = history['val_accuracy'][-1]
        
        # 獲取驗證準確率的最大值及其epoch
        best_val_acc = max(history['val_accuracy'])
        best_val_acc_epoch = history['val_accuracy'].index(best_val_acc) + 1
        
        # 計算訓練和驗證曲線之間的平均差距
        mean_train_val_gap = np.mean(np.array(history['accuracy']) - np.array(history['val_accuracy']))
        
        report_data.append([
            result["name"],
            f"{result['train_acc']:.4f}",
            f"{result['test_acc']:.4f}",
            f"{result['overfitting_gap']:.4f}",
            f"{final_val_acc:.4f}",
            f"{best_val_acc:.4f}",
            f"{best_val_acc_epoch}",
            f"{mean_train_val_gap:.4f}"
        ])
    
    # 按測試集準確率排序
    report_data.sort(key=lambda x: float(x[2]), reverse=True)
    
    headers = [
        "模型", 
        "訓練集準確率", 
        "測試集準確率", 
        "過擬合差距",
        "最終驗證準確率",
        "最佳驗證準確率",
        "最佳驗證Epoch",
        "訓練-驗證平均差距"
    ]
    
    try:
        # 使用pandas和tabulate顯示詳細報告
        detailed_df = pd.DataFrame(report_data, columns=headers)
        print(tabulate(detailed_df, headers='keys', tablefmt='pipe', showindex=False))
        
        # 保存詳細報告到CSV
        detailed_df.to_csv('detailed_regularization_results.csv', index=False, float_format='%.4f')
        print("\n詳細報告已保存到 'detailed_regularization_results.csv'")
        
    except ImportError:
        # 基本格式化輸出
        print("{:<35} {:<12} {:<12} {:<12} {:<12} {:<12} {:<12} {:<12}".format(*headers))
        print("-" * 120)
        
        for row in report_data:
            print("{:<35} {:<12} {:<12} {:<12} {:<12} {:<12} {:<12} {:<12}".format(*row))

# 生成詳細報告
generate_detailed_report()
```
